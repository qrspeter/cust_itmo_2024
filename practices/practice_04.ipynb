## Problem 

Gauss function

$f(x)=\frac{1}{\sqrt(2\pi) s} exp \left[-\frac{1}{2}\left( \frac{x-m}{s} \right)^2 \right]$


Convert yesterday's code (see below) from scalar to NumPy array arguments

```python
import math

print(math.pi)

def gauss(x, m=1, s=1):
    first = 1/math.sqrt(2*math.pi)
    second = -0.5*((x - m)/s)**2
    result = first * math.exp(second)
    
    return result

x = 5
m = 5
s = 3
print(gauss(x, m, s))

x_range = list(range(-4, 7))

y_data = []
for x in x_range:
    y_data.append(gauss(x))
    print(f'{y_data[-1]:.5e}')
```

```python
import numpy as np

def gauss(x, m=1, s=1):
    first = 1/math.sqrt(2*math.pi)
    second = -0.5*((x - m)/s)**2
    result = first * np.exp(second)
    
    return result

x = np.arange(1,10)

print(x)
print(gauss(x))
```
