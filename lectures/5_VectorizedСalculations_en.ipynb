### Comparizon NumPy with Built-in math for array.

The single operations have the same speed (https://stackoverflow.com/questions/3650194/are-numpys-math-functions-faster-than-pythons):


````python
import numpy as np
import timeit
 
x = np.array([-1.56]) #np.random.standard_normal(10000)
 
def pure_abs():
    return abs(x)
 
def numpy_abs():
    return np.abs(x)
 
n = 10000
 
t1 = timeit.timeit(pure_abs, number = n)
print('Pure Python abs:', t1)
t2 = timeit.timeit(numpy_abs, number = n)
print('Numpy abs:', t2)
````

And much faster for arrays (https://stackoverflow.com/questions/36526708/comparing-python-numpy-numba-and-c-for-matrix-multiplication):

````python
import numpy as np
import timeit

def dot_py(A, B):
    m, n = A.shape
    p = B.shape[1]

    C = np.zeros((m,p))

    for i in range(0,m):
        for j in range(0,p):
            for k in range(0,n):
                C[i,j] += A[i,k]*B[k,j] 
    return C

def dot_np(A, B):
    C = np.dot(A,B)
    return C

m  = 100
A = np.random.random((m,m))
B = np.random.random((m,m))

n = 10

#c = dot_py(A, B)
t1 = timeit.timeit(lambda: dot_py(A, B), number = n)
print('Pure Python dot:', t1)
t2 = timeit.timeit(lambda: dot_np(A, B), number = n)
print('Numpy dot:', t2)
````

### Vectorized operation in microprocessors

SIMD - single instruction, multiple data (https://en.wikipedia.org/wiki/Single_instruction,_multiple_data)

MMX - MMX is a single instruction, multiple data (SIMD) instruction set architecture designed by Intel, introduced on January 8, 1997 (https://en.wikipedia.org/wiki/MMX_(instruction_set))

SSE - single instruction, multiple data (SIMD) instruction set extension to the x86 architecture, designed by Intel and introduced in 1999 (https://en.wikipedia.org/wiki/Streaming_SIMD_Extensions)

AVX - Advanced Vector Extensions (AVX) are SIMD extensions to the x86 instruction set architecture for microprocessors from Intel and Advanced Micro Devices (AMD). They were proposed by Intel in March 2008 (https://en.wikipedia.org/wiki/Advanced_Vector_Extensions).
