### Logarithmic scale

Functions:
* Linear - $ax$
* Power - $x^a$
* Exponential - $a^x$
* Logarithmic - $ln(x)$



Exponential decay and exponential growth:

````python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

koeff = [-1, 1]

for k in koeff:
    x = np.linspace(0, 10, 100)
    y = np.exp(k * x)
    
    figure, axis = plt.subplots(1, 2, figsize=(8, 3)) 
    
    axis[0].plot(x, y, c="red", label=f'k={k}')
    axis[0].set_title("Linear scale") 
    axis[0].set(xlabel='Time', ylabel='Intensity')
    axis[0].legend()
      
    axis[1].plot(x, y, c="red")
    axis[1].set_title("Logarithmic scale") 
    axis[1].set_yscale('log')
    axis[1].set(xlabel='Time')
    
    plt.show() 
````

Double logarithmic scale (log-log):

````python
plt.figure(figsize = (6,4))
x = list(range(-10, 10))
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]
plt.plot(x, y1, 'k-', label = 'square')
plt.plot(x, y2, 'r:', label = 'cubic')
plt.title('Double logarithmic scale')
plt.yscale('log')
plt.xscale('log')

plt.legend()
plt.show()
````

### Multiple graphs in one picture using subplots

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

```python
matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, width_ratios=None, height_ratios=None, subplot_kw=None, gridspec_kw=None, **fig_kw)
```

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 10)
y1 = x ** 2
y2 = x ** 3

plt.figure(figsize = (6, 4))
# 2 in a row, 1 in columns, 1st
plt.subplot(1,2, 1)
plt.plot(x, y1, 'g', linewidth=2)

# 2 in a row, 1 in columns, 2nd
plt.subplot(1, 2, 2)
plt.plot(x, y2, 'r')

plt.show()
````

Or using row/column indexes:

## Using your own color scheme

Sometimes you want, for example, the order of colors to not change from one drawing to another (different Python distributions may choose them differently).

````python
samples = 5
x = np.linspace(0, 10, 20)
colors = [plt.cm.jet(float(i)/samples) for i in range(samples)]

for i in range(samples):
    col = np.expand_dims(np.array(colors[i]), axis=0)
    plt.plot(x, x**i, c=col, label=f'x$^{i}$')

plt.legend()
plt.yscale('log')
plt.show()
````

### TeX

To use TeX notation in Jupyter Notebook you need:
* The r string specifier (not always)
* Surround the equation with a pair of \$ characters
* Nothing else (built-in support)

Elements of notation:
* ^ - upper case (_a will be $_a$)
* _ - lower case (^a will be $^a$)
* \ - precedes commands, special characters, Greek letters.
* Greek letters \alpha - $\alpha$
* {} - grouping with curly brackets, for example if the equation in the power is (a^{b+c} will be $a^{b+c}$)
* \frac{a}{b} - division $\frac{a}{b}$
* \left( \right) - brackets are adjusted according to the size of the contents - \left(\frac{a}{b}\right) will be $\left(\frac{a}{b}\right)$
* \int - integral, and \int_a^b will be $\int_a^b$

````python
x = list(range(-10, 10))
y = [i**2 for i in x]
plt.plot(x, y, 'ko', label = r'$x_3$ function')
plt.legend()
plt.show()
````
